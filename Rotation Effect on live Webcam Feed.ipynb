```python
#OpenCV Mini Project - Rotation Effect on live Webcam Feed
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time

def main():
    imgpath="C:\\Users\\DELL\\Desktop\\Machine Learning A-Z Template Folder\\standard_test_images\\"
    
    imgpath1=imgpath+"cup.tiff"
    
    img=cv2.imread(imgpath1,1)
   
    
    rows,column,channel=img.shape
    angle = 360
    while True:
        if angle==0:
            angle=360
            
        R=cv2.getRotationMatrix2D((column/2,rows/2),angle,0.5)
        print(R)
    
        output=cv2.warpAffine(img,R,(column,rows))
    
        cv2.imshow('Rotated Image', output)
        angle=angle-1
        time.sleep(0.01)
        
        if cv2.waitKey(1) == 27:
            break
        
        
    cv2.destroyWindow('rotated image')
    
if __name__=='__main__':
    main()
```
